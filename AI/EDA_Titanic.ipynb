# Data Visualization

## Univariate Analysis

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def numeric_plot_side_by_side(data, column):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # Histogram
    sns.histplot(data[column], kde=False, ax=axes[0])
    axes[0].set_title('Histogram')

    # Box Plot
    sns.boxplot(x=data[column], ax=axes[1])
    axes[1].set_title('Box Plot')

    # Density Plot
    sns.kdeplot(data[column], ax=axes[2])
    axes[2].set_title('Density Plot')

    plt.tight_layout()
    plt.show()
```

```python
numeric_plot_side_by_side(df, "Age")
```

```python
numeric_plot_side_by_side(df, "Fare")
```

## Bivariate Analysis

```python
sns.pairplot(df)
```

```python
def plot_correlation_heatmap(df):
    # Select only numeric columns
    numeric_df = df.select_dtypes(include=[float, int])

    # Calculate the correlation matrix
    corr = numeric_df.corr()

    # Create a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5, linecolor='white')
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
plot_correlation_heatmap(df)
```

## Multivariate Analysis

```python
def plot_3d_visualizations(df, columns):
    if len(columns) != 3:
        raise ValueError("The list of columns must contain exactly three column names.")

    x_col, y_col, z_col = columns

    # Ensure the columns are numeric
    df[x_col] = pd.to_numeric(df[x_col], errors='coerce')
    df[y_col] = pd.to_numeric(df[y_col], errors='coerce')
    df[z_col] = pd.to_numeric(df[z_col], errors='coerce')

    # Drop rows with NaN values in any of the specified columns
    df = df.dropna(subset=columns)

    # Prepare data for 3D scatter plot
    x_scatter = df[x_col]
    y_scatter = df[y_col]
    z_scatter = df[z_col]

    # Prepare data for 3D surface plot
    x_surface = np.linspace(df[x_col].min(), df[x_col].max(), 100)
    y_surface = np.linspace(df[y_col].min(), df[y_col].max(), 100)
    x_surface, y_surface = np.meshgrid(x_surface, y_surface)
    z_surface = np.sin(np.sqrt(x_surface**2 + y_surface**2))

    # Prepare data for 3D contour plot
    z_contour = np.cos(np.sqrt(x_surface**2 + y_surface**2))

    # Create subplots
    fig = plt.figure(figsize=(18, 6))

    # 3D Scatter Plot
    ax1 = fig.add_subplot(1, 3, 1, projection='3d')
    scatter = ax1.scatter(x_scatter, y_scatter, z_scatter, c=z_scatter, cmap='viridis')
    ax1.set_title('3D Scatter Plot')
    ax1.set_xlabel(x_col)
    ax1.set_ylabel(y_col)
    ax1.set_zlabel(z_col)
    fig.colorbar(scatter, ax=ax1, shrink=0.5, aspect=5)

    # 3D Surface Plot
    ax2 = fig.add_subplot(1, 3, 2, projection='3d')
    surface = ax2.plot_surface(x_surface, y_surface, z_surface, cmap='coolwarm')
    ax2.set_title('3D Surface Plot')
    ax2.set_xlabel(x_col)
    ax2.set_ylabel(y_col)
    ax2.set_zlabel('Z')
    fig.colorbar(surface, ax=ax2, shrink=0.5, aspect=5)

    # 3D Contour Plot
    ax3 = fig.add_subplot(1, 3, 3, projection='3d')
    contour = ax3.contour3D(x_surface, y_surface, z_contour, 50, cmap='viridis')
    ax3.set_title('3D Contour Plot')
    ax3.set_xlabel(x_col)
    ax3.set_ylabel(y_col)
    ax3.set_zlabel('Z')

    plt.tight_layout()
    plt.show()
```

```python
plot_3d_visualizations(df, ['Age', 'Fare', 'Survived'])
```

```python
plot_3d_visualizations(df, ['Ticket', 'Fare', 'Survived'])
```
