### Transformada Inversa para la Distribuci√≥n Uniforme

```python

import numpy as np
import matplotlib.pyplot as plt

# Par√°metros
a, b = 2, 5  # intervalo de la distribuci√≥n uniforme
n = 1000  # n√∫mero de muestras

# Generaci√≥n de variables aleatorias
u = np.random.rand(n)
x = a + (b - a) * u  # Transformada inversa

# Visualizaci√≥n
plt.hist(x, bins=30, density=True, alpha=0.6, color="skyblue", edgecolor="black")
plt.title("Distribuci√≥n Uniforme usando Transformada Inversa")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Transformada Inversa para la Distribuci√≥n Exponencial

```python

import numpy as np
import matplotlib.pyplot as plt

# Par√°metro
lmbda = 1  # tasa de la distribuci√≥n exponencial
n = 1000  # n√∫mero de muestras

# Generaci√≥n de variables aleatorias
u = np.random.rand(n)
x = -np.log(1 - u) / lmbda  # Transformada inversa

# Visualizaci√≥n
plt.hist(x, bins=30, density=True, alpha=0.6, color="lightcoral", edgecolor="black")
plt.title("Distribuci√≥n Exponencial usando Transformada Inversa")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Transformada Inversa para la Distribuci√≥n de Bernoulli

```python

import numpy as np
import matplotlib.pyplot as plt

# Par√°metro
p = 0.3  # probabilidad de √©xito
n = 1000  # n√∫mero de muestras

# Generaci√≥n de variables aleatorias
u = np.random.rand(n)
x = (u < p).astype(int)  # Transformada inversa

# Visualizaci√≥n
plt.hist(x, bins=2, density=True, alpha=0.6, color="mediumseagreen", edgecolor="black", rwidth=0.8)
plt.title("Distribuci√≥n de Bernoulli usando Transformada Inversa")
plt.xticks([0, 1], ["Fracaso", "√âxito"])
plt.xlabel("Resultado")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```
