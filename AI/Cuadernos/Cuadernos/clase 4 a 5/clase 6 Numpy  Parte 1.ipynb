#### Construir arrays a partir de una lista simple y una lista anidada usando tipos de datos numéricos distintos para cada una


```python
import numpy as np
```

```python
arr1 = np.array([1,5,6,3,5,6,2])
arr1
```

```python
arr1.dtype
```

```python
arr2 = np.array([1,5,6,3,5,6,2], dtype=np.int8)
arr2
```

```python
arr3 = np.array([1,5,6,3,5,6,2], dtype=np.float32)
arr3
```

```python
arr4 = np.array([[4,8,6],[2,3,5],[7,5,7]])
arr4, arr4.dtype, arr4.ndim, arr4.shape
```

```python
arr5 = np.array([[4,8,6],[2,3,5],[7,5,7]], dtype=np.float)
arr5, arr5.dtype
```
