# Distribucion Normal

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

n = 100000  # número de muestras
uniform1 = np.random.uniform(0, 1, n)
uniform2 = np.random.uniform(0, 1, n)

# Aplicamos la transformación de Box-Muller para obtener una distribución normal estándar
normal_dist = np.sqrt(-2 * np.log(uniform1)) * np.cos(2 * np.pi * uniform2)

# Graficamos el resultado
sns.histplot(normal_dist)
plt.title("Distribución Normal (Transformación de Box-Muller)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

## ¿Qué es la Transformación de Box-Muller?
La transformación de Box-Muller es un método matemático que permite generar números aleatorios con una distribución normal (o gaussiana) a partir de números aleatorios que siguen una distribución uniforme. En otras palabras, es una técnica que nos ayuda a convertir números aleatorios "comunes" (uniformes) en números aleatorios que siguen la famosa curva de campana.

## ¿Cómo funciona?
La transformación de Box-Muller se basa en una propiedad matemática que relaciona la distribución uniforme con la distribución normal. En esencia, toma dos números aleatorios uniformes entre 0 y 1, y los transforma en dos números aleatorios independientes que siguen una distribución normal estándar (media 0 y desviación estándar 1).

La fórmula matemática detrás de esta transformación es un poco compleja, pero la idea principal es que se aplican ciertas operaciones matemáticas (raíces cuadradas, logaritmos, funciones trigonométricas) a los números uniformes para obtener los números normales.

# Distribucion binomial

```python
import numpy as np
import matplotlib.pyplot as plt

n = 5           # Tamaño del lote
p_exito = 0.5  # Probabilidad de éxito en cada ensayo (encontrar una pieza defectuosa)
num_simulaciones = 10000  # Número de simulaciones para obtener una buena aproximación

# Aplicar el método de la convolución: sumamos n variables de Bernoulli(p)
binomial_convolucion = np.sum(np.random.binomial(1, p_exito, (num_simulaciones, n)), axis=1)

# Graficamos el resultado
plt.hist(binomial_convolucion, bins=range(n+2), density=True, color="skyblue", edgecolor="black", align='left')
plt.title("Distribución Binomial (Método de la Convolución)")
plt.xlabel("Número de Éxitos")
plt.ylabel("Densidad")
plt.show()
```

`np.random.binomial(1, p_exito, (num_simulaciones, n))`

Esta línea genera una matriz de variables aleatorias de Bernoulli, donde:

`1` es el número de ensayos de cada variable Bernoulli. Con un solo ensayo por fila, el resultado es un éxito (1) o un fracaso (0).

`p_exito` es la probabilidad de éxito (o que se obtenga un "1") en cada ensayo.

`(num_simulaciones, n)` indica la forma de la matriz generada, en la que:

- `num_simulaciones` es la cantidad de simulaciones independientes (filas).
- `n` es el número de variables Bernoulli generadas en cada simulación (columnas).

--- 

**Por ejemplo**, si num_simulaciones es 10000 y n es 5, np.random.binomial(1, p_exito, (10000, 5)) generará una matriz de 10,000 filas y 5 columnas. Cada elemento de la matriz será 0 o 1, con 1 representando un "éxito" según la probabilidad p_exito.

---

Generación de una Distribución Binomial mediante Convolución
El método de la convolución genera una distribución binomial sumando los resultados de estas variables de Bernoulli en cada simulación. Al aplicar np.sum(..., axis=1), estamos obteniendo el número total de éxitos en cada fila (o simulación), produciendo una lista de valores binomiales.
