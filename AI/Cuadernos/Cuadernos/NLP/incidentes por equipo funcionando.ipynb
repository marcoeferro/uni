```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X)
X_train_counts.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tfidf, y)
```

```python
docs_new = ['REVISAR AIREADOR VERTICAL LAGUNA N°2', 'REVISAR FALTA DE AGUA EN SECTOR ADMINISTRACIÓN', 'MANTENIMIENTO MENSUAL EN STV']
X_new_counts = count_vect.transform(docs_new)
X_new_tfidf = tfidf_transformer.transform(X_new_counts)

predicted = clf.predict(X_new_tfidf)

for doc, category in zip(docs_new, predicted):
    print('%r => %s' % (doc, category))
```

```python
from sklearn.pipeline import Pipeline
text_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB()),
])
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
# Split the dataset in training and test set:
docs_train, docs_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5)
```

```python
text_clf.fit(docs_train, y_train)
```

```python
import numpy as np
predicted = text_clf.predict(docs_test)
np.mean(predicted == y_test)
```

```python
from sklearn.linear_model import SGDClassifier
text_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', SGDClassifier(loss='hinge', penalty='l2',
                          alpha=1e-3, random_state=42,
                          max_iter=5, tol=None)),
])

text_clf.fit(docs_train, y_train)
predicted = text_clf.predict(docs_test)
np.mean(predicted == y_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, predicted))




metrics.confusion_matrix(y_test, predicted)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Datos de ejemplo
textos = list(problemas_por_serie_PSALPLP5.index)

# Preprocesamiento de textos
preprocesador = CountVectorizer()
texto_preprocesado = preprocesador.fit_transform(textos)

# Word embeddings utilizando matriz TF-IDF
word_embeddings = texto_preprocesado.toarray()

# Cálculo de similitud coseno
similitud = cosine_similarity(word_embeddings)

# Buscar palabras similares a una palabra de consulta
consulta = "agregar"
if consulta in preprocesador.vocabulary_:
    indice_consulta = preprocesador.vocabulary_[consulta]
    similitudes_consulta = similitud[indice_consulta]

    # Obtener las palabras más similares
    top_similares = np.argsort(-similitudes_consulta)[:3]  # Las 3 palabras más similares
    palabras_similares = [palabra for palabra, indice in preprocesador.vocabulary_.items() if indice in top_similares]

    print(f"Palabras similares a '{consulta}': {palabras_similares}")
else:
    print(f"'{consulta}' no está presente en el vocabulario.")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from gensim.models import Word2Vec

# Datos de ejemplo
textos = list(problemas_por_serie_PSALPLP5.index)

# Preprocesamiento de textos
preprocesador = CountVectorizer()
texto_preprocesado = preprocesador.fit_transform(textos)
vocabulario = list(preprocesador.vocabulary_.keys())

# Entrenamiento del modelo Word2Vec
sentences = [texto.split() for texto in textos]
modelo = Word2Vec(sentences, min_count=1, vector_size=100)

# Construcción de la matriz de word embeddings
word_embeddings = np.zeros((len(vocabulario), 100))
for i, palabra in enumerate(vocabulario):
    if palabra in modelo.wv:
        word_embeddings[i] = modelo.wv[palabra]

# Cálculo de similitud coseno
similitud = cosine_similarity(word_embeddings)

# Buscar palabras similares a una palabra de consulta
consulta = "agregar"
if consulta in preprocesador.vocabulary_:
    indice_consulta = preprocesador.vocabulary_[consulta]
    similitudes_consulta = similitud[indice_consulta]

    # Obtener las palabras más similares
    top_similares = np.argsort(-similitudes_consulta)[:3]  # Las 3 palabras más similares
    palabras_similares = [vocabulario[indice] for indice in top_similares]

    print(f"Palabras similares a '{consulta}': {palabras_similares}")
else:
    print(f"'{consulta}' no está presente en el vocabulario.")
```
