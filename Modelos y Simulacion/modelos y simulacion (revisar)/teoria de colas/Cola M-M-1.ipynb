importacion de modulos

```python
import numpy as np
import numpy.random as rnd

import pandas as pd

```

definicion de funciones auxiliares

```python
def acumular_array(array):
    acumulado = 0
    array_acumulado = []
    for n in array :
        acumulado = n + acumulado
        array_acumulado.append(acumulado)
    return array_acumulado
```

codigo de la simulacion

```python
llegadas = 10 #llegan 10 clientes x hora
capacidad_servicio = 12 #podemos atender un max. de 12 clientes por hora
#seccion llegadas
num_clientes = 25 
df = pd.DataFrame()
pseudo1 = [rnd.uniform(0,1) for _ in range(num_clientes)]
tiempo_entre_llegadas = [(-np.log(1-c)/llegadas)*60 for c in pseudo1]
momento_de_llegada = acumular_array(tiempo_entre_llegadas)

#seccion servicios
pseudo2 = [rnd.uniform(0,1) for _ in range(num_clientes)]
tiempo_de_servicio = [(-np.log(1-c)/capacidad_servicio)*60 for c in pseudo2]
tiempo_inicio_servicio = np.zeros(num_clientes)
tiempo_terminacion_servicio = np.zeros(num_clientes)
tiempo_espera = np.zeros(num_clientes)

for i in range(num_clientes):
    if i == 0:
        tiempo_inicio_servicio[i] = momento_de_llegada[i]
    else:
        tiempo_inicio_servicio[i] = max(momento_de_llegada[i], tiempo_terminacion_servicio[i-1])
    tiempo_terminacion_servicio[i] = tiempo_inicio_servicio[i] + tiempo_de_servicio[i]

for i in range(num_clientes):
    aux = tiempo_inicio_servicio[i] - momento_de_llegada[i]
    if aux < 0 : aux = 0
    tiempo_espera[i] = aux


df.insert(0, 'tiempo entre llegadas' , tiempo_entre_llegadas)
df.insert(1, 'momento de llegada', momento_de_llegada)
df.insert(2, 'tiempo de inicio de servicio', tiempo_inicio_servicio)
df.insert(3, 'tiempo de espera', tiempo_espera)
df.insert(4, 'tiempo de servicio', tiempo_de_servicio)
df.insert(5, 'tiempo de terminacion de servicio', tiempo_terminacion_servicio)
df
```

```python
total_tiempo_espera = sum(tiempo_espera)
print(total_tiempo_espera)
```
