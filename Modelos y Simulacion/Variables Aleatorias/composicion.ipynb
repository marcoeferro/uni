# Distribución Mixta de Dos Exponenciales
Este notebook genera una distribución compuesta de dos distribuciones exponenciales con diferentes tasas de decaimiento (lambda) y visualiza el histograma resultante.
Se utiliza una variable aleatoria uniforme para decidir cuál de las dos distribuciones asignar a cada muestra.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las distribuciones
lambda_values = [0.5, 1.5]  # lambdas para las distribuciones
mixing_prob = 0.3  # Probabilidad de seleccionar la primera distribución

# Generar muestras
num_samples = 10000
uniform_samples = np.random.uniform(0, 1, num_samples)
mixed_samples = np.where(
    uniform_samples < mixing_prob,
    np.random.exponential(1 / lambda_values[0], num_samples),
    np.random.exponential(1 / lambda_values[1], num_samples)
)

# Graficar histograma
plt.hist(mixed_samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title('Distribución Mixta de dos Exponenciales')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```

### Explicación del Código
Este código genera una muestra de 10,000 valores que sigue una **distribución mixta de dos distribuciones exponenciales** con tasas de decaimiento (`lambda`) diferentes.
- **Paso 1**: Se definen dos valores `lambda`, uno para cada distribución exponencial, y una probabilidad `mixing_prob` que define la probabilidad de que un valor provenga de la primera distribución.
- **Paso 2**: Se generan valores uniformes `uniform_samples` para determinar cuál de las dos distribuciones se usa para cada muestra.
- **Paso 3**: Dependiendo del valor en `uniform_samples`, cada muestra se toma de la primera o segunda distribución exponencial.
- **Paso 4**: Se grafica un histograma que muestra la distribución resultante, combinando las dos exponenciales.


### Ejemplo de Uso
Este notebook es útil para ilustrar cómo combinar distribuciones para obtener una **distribución mixta**. Esta técnica puede aplicarse en simulaciones donde los datos provienen de múltiples fuentes o categorías con distribuciones diferentes.
