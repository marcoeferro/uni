```python
import numpy as np
import matplotlib.pyplot as plt
```

definimos los parametros de las distribuciones

```python
lambda_1 = 0.5 #lambda para la primer distribucion
lambda_2 = 1.5 #lambda para la segunda distribucion
limite = 0.3
```

generamos las muestras y evaluamos la funcion con un lambda distinto dependiendo del valor

```python
n= 10000 # numero de muestras
U = np.random.uniform(0,1,n)  # Variables uniformes para decidir la elección
X = np.where(U < limite, 
            np.random.exponential(1 / lambda_1, n), 
            np.random.exponential(1 / lambda_2, n))
```

graficamos un histograma

```python
# Graficar el histograma
plt.hist(X, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title("Distribución Mixta de dos Exponenciales")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```
