### Transformada Inversa para la Distribución Uniforme

```python

import numpy as np
import matplotlib.pyplot as plt

# Parámetros
a, b = 2, 5  # intervalo de la distribución uniforme
n = 1000  # número de muestras

# Generación de variables aleatorias
u = np.random.rand(n)
x = a + (b - a) * u  # Transformada inversa

# Visualización
plt.hist(x, bins=30, density=True, alpha=0.6, color="skyblue", edgecolor="black")
plt.title("Distribución Uniforme usando Transformada Inversa")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Transformada Inversa para la Distribución Exponencial

```python

import numpy as np
import matplotlib.pyplot as plt

# Parámetro
lmbda = 1  # tasa de la distribución exponencial
n = 1000  # número de muestras

# Generación de variables aleatorias
u = np.random.rand(n)
x = -np.log(1 - u) / lmbda  # Transformada inversa

# Visualización
plt.hist(x, bins=30, density=True, alpha=0.6, color="lightcoral", edgecolor="black")
plt.title("Distribución Exponencial usando Transformada Inversa")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Transformada Inversa para la Distribución de Bernoulli

```python

import numpy as np
import matplotlib.pyplot as plt

# Parámetro
p = 0.3  # probabilidad de éxito
n = 1000  # número de muestras

# Generación de variables aleatorias
u = np.random.rand(n)
x = (u < p).astype(int)  # Transformada inversa

# Visualización
plt.hist(x, bins=2, density=True, alpha=0.6, color="mediumseagreen", edgecolor="black", rwidth=0.8)
plt.title("Distribución de Bernoulli usando Transformada Inversa")
plt.xticks([0, 1], ["Fracaso", "Éxito"])
plt.xlabel("Resultado")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```
