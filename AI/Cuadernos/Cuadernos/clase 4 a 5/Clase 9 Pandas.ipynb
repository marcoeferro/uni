Importar el archivo "GoldPrice.csv" y realizar los siguientes cálculos:

Determinar el promedio de precios por año.

Calcular el año con mayor promedio y el año con menor promedio.

Devolver los resultados bajo el formato:

Mayor precio promedio Año:xxx Precio:xxxx

Menor precio promedio Año:xxxx Precio: xxxx

```python
#cargamos nuestro archivo de delitos en memoria
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
df = pd.read_csv("GoldPrice.csv")
df
```

```python
df.dtypes
```

```python
#1. Convierto a datetime la fecha
df["Date"] = pd.to_datetime(df["Date"])
df.dtypes
```

```python
df["Año"] = df["Date"].dt.year
```

```python
df
```

```python
#Promedios por año
df[["Año","Price"]].groupby("Año").mean().head()  # si le saco el head() , me los muestra a todos
```

```python
#Año  minimo promedio
año = int(df[["Año","Price"]].groupby("Año").mean().idxmin())
precio = float(df[["Año","Price"]].groupby("Año").mean().min())
print("Año: ", año)
print("Precio: ", precio)
```

```python
#Año máximo promedio
año = int(df[["Año","Price"]].groupby("Año").mean().idxmax())
precio = float(df[["Año","Price"]].groupby("Año").mean().max())
print("Año: ", año)
print("Precio: ", precio)
```
