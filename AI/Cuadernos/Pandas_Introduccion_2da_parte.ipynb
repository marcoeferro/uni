```python
data = pd.read_excel ('condicion_habitacional.xlsx', sheet_name = "Cuadro 1" )
data
```

```python
data_ordenado_simple = data_ordenado[['Jurisdicción', 'total']]
```

```python
data_ordenado_simple
```

```python
data_ordenado_simple.columns
```

```python
plt.figure()
data_ordenado_simple.plot(x="Jurisdicción", y="total", kind="bar", title = "Puntaje Total por Provincia", ylabel = "Total" , figsize=(9,6))
plt.savefig("puntaje por provincia.jpg", bbox_inches='tight')
```

# Nueva sección

```python
total = data_ordenado_simple['total']
provincia = data_ordenado_simple['Jurisdicción']

plt.grid(True)
plt.plot(provincia,total, marker="o", label='Total de Puntos')

plt.legend(loc='upper center')
plt.title('Evolucion de Indicadores Socio Economicos')
plt.xlabel('Provincias')
plt.ylabel('Puntos Totales')
plt.gcf().set_size_inches(10, 9)
plt.xticks(rotation=80)
plt.show()
```

```python
corr_data = data.corr(method='pearson')

plt.figure(figsize=(9, 9))
sns.heatmap(corr_data, annot=True, fmt=".2f") #, cmap="crest"
plt.show()
```

```python
pisos_con_revestimiento = data['Pisos con revestimiento']
celular_con_internet = data['Celular con internet']
provincia = data['Jurisdicción']

plt.grid(True)
plt.plot(provincia,pisos_con_revestimiento, marker="o", label='Pisos con revestimiento')
plt.plot(provincia,celular_con_internet, marker=".", label='Celular con internet')

plt.legend(loc='upper right')
plt.title('Indicadores Habitacionales Total Pais 2022')
plt.xlabel('Provincias')
plt.ylabel('Puntos')
plt.gcf().set_size_inches(10, 9)
plt.xticks(rotation=90)
plt.show()
```

```python
data_cordoba = pd.read_excel ('condicion_habitacional_concat.xlsx', sheet_name = "Cuadro 1.6" )
data_cordoba
```

```python
# inserta en la primera columna ( 0 ), la Columna 'Provincia' con el valor ' Cordoba'
data_cordoba.insert(0, 'Provincia', "Cordoba")
```

```python
data_cordoba
```

```python
# Alternativa 1. utilizando iloc . Detecto fila y columna a reemplazar en este caso Capital , ver en la celda de arriba y le doy nombre que quiero Cordoba Capitaaaal
data_cordoba.iloc[[2],[1]] = "Cordoba Capitaaaal"
data_cordoba
```

```python
# Alternativa 2. con replace. Busca en el dataframe la palabra Cordoba Capitaaaal y la reemplaza por Cordoba Capital
data_cordoba.replace("Cordoba Capitaaaal", "Cordoba Capital", inplace=True)
data_cordoba
```

```python
data_corrientes = pd.read_excel ('condicion_habitacional_concat.xlsx', sheet_name = "Cuadro 1.7" )
data_corrientes
```

```python
# inserta en la primera columna ( 0 ), la Columna 'Provincia' con el valor ' Corrientes'
data_corrientes.insert(0, 'Provincia', "Corrientes")
data_corrientes
```

```python
data_corrientes.replace("Capital", "Corrientes Capital", inplace=True)
data_corrientes
```

```python
# Ahora resta unir los dos dataframe de Cordoba y Corrientes
data_concatenado = pd.concat([data_cordoba , data_corrientes], sort = False)
data_concatenado
```
