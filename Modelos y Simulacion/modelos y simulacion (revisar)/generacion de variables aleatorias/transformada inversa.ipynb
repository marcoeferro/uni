```python
# Paquete Numpy
import numpy as np

# Generación de números pseudoaleatorios
from numpy import random as rd

# Módulo para las gráficas
from matplotlib import pyplot as plt

# Análisis de datos
import pandas as pd
from pandas import DataFrame as df
```

1. definimos la funcion de densidad, obtenemos la acumulada y su inversa

```python
import sympy as sp

# Definir la variable y la función
x = sp.symbols('x')
f = 3*x**2  # Función f(x) = 3x^2
#f = 2*x

# Definir la variable para la inversa
y = sp.symbols('y')
# Igualamos f(x) a y y despejamos x para encontrar la inversa
F_acum = sp.integrate(f,x)
inverse_f = sp.solve(F_acum - y, x)
```

2. Obtemenos una distribucion uniforme de numeros pseudoaleatorios R y evaluamos inverse_F en funcion de R

```python
R = np.random.random(10000)
F_inv = sp.lambdify(y, inverse_f[0], modules="numpy")
print(F_inv(R))
X = F_inv(R)
```

3.graficamos

```python
# Graficando 
plt.plot(R, X, '.')

#Etiquetas
plt.ylabel('R')
plt.xlabel('x = F_inv(R)')
plt.title ('Valores generados con transformada inversa')
plt.show()
```
