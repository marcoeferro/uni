```python
import numpy as np
import matplotlib.pyplot as plt
```

Metodo de la Convolucion para generacion de variables aleatorias con distribucion normal

```python
n = 100000  # número de muestras
uniform1 = np.random.uniform(0, 1, n)
uniform2 = np.random.uniform(0, 1, n)

# Aplicamos la transformación de Box-Muller para obtener una distribución normal estándar
normal_dist = np.sqrt(-2 * np.log(uniform1)) * np.cos(2 * np.pi * uniform2)

# Graficamos el resultado
plt.hist(normal_dist, bins=50, density=True, color="green", edgecolor="black")
plt.title("Distribución Normal (Transformación de Box-Muller)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

Distribucion binomial

```python
n = 5       # tamaño del lote
p_exito = 0.03     # probabilidad de éxito en cada ensayo (encontrar una pieza defectuosa)

binomial = np.random.binomial(n=n,p=p_exito,size=n)

# Graficamos el resultado
plt.hist(binomial, bins=50, density=True, color="skyblue", edgecolor="black")
plt.title("Distribución Binomial (Método de la Convolución)")
plt.xlabel("Número de Éxitos")
plt.ylabel("Densidad")
plt.show()

```
