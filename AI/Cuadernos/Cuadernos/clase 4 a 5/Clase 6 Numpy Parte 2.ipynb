```python
def farenheit(c):
    return c*1.8 + 32
```

Hacer una función que tome dos arrays (peso y altura) y calcule el indice de masa corporal

```python
def imc(peso, altura):
    return peso/altura**2
```

```python
altura = np.array([1.78, 1.87, 1.60, 1.96])
peso = np.array([78, 87, 60, 96])
imc(peso,altura)
```
